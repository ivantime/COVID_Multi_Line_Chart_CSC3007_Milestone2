# Combine CSVs From John Hopkins University to 1 Json File (data.json)

```python
import os,pandas as pd,numpy as np, json
dfArray=[]
dates=[]
folderPath="C:\\Users\\I\\Desktop\\covid_multi_line_chart_CSC3007_milestone2\\jhu_csvs"
directory = os.path.join(folderPath)
count=1
for root,dirs,files in os.walk(directory):
    for file in files:
        if (file.endswith(".csv") and count<7):
            df = pd.read_csv(folderPath+"\\"+file)
            try:
                df.drop(columns=df.columns[[0,1,2,4,5,6,9,10,11,12,13]],errors='ignore',axis=1,inplace=True)
            except:
                pass
            df.rename(columns={"Country_Region":"Country"},errors='ignore',inplace=True)
            df.rename(columns={"Country/Region":"Country"},errors='ignore',inplace=True)
            df=df.groupby("Country",as_index=False).sum()
            dates.append(file.split(".")[0])
            dfArray.append(df)
            count+=1
```

```python
len(dfArray)
```

```python
out={}
for i in range(0,len(dates)):
    out[dates[i]]= dfArray[i].groupby("Country").last().to_dict(orient='index')
with open("data\\country_data.json", "w") as outfile:
    json.dump(out, outfile, indent=4, sort_keys=False)
```
